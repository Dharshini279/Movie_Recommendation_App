```python
import ast
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies_df['genres']= movies_df['genres'].apply(convert)
movies_df['keywords'] = movies_df['keywords'].apply(convert)
movies_df.head()
```

```python
def convert3(obj):
    L=[]
    c=0
    for i in ast.literal_eval(obj):
        if c!=3:
            L.append(i['name'])
            c+=1
        else:
            break
    return L
```

```python
movies_df['cast']=movies_df['cast'].apply(convert3)
```

```python
movies_df.head()
```

```python
def fetch_dir(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies_df['crew']=movies_df['crew'].apply(fetch_dir)
```

```python
movies_df['overview']=movies_df['overview'].apply(lambda x:x.split())
```

```python
movies_df['genres']=movies_df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies_df['keywords']=movies_df['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies_df['cast']=movies_df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies_df['crew']=movies_df['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies_df['tags']=movies_df['overview']+movies_df['genres']+movies_df['keywords']+movies_df['cast']+movies_df['crew']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vec=cv.fit_transform(new_df['tags']).toarray()
```

```python
vec[0]
```

```python
len(cv.get_feature_names_out())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vec)
```

```python
sim=cosine_similarity(vec)
```

```python
sim[0]
```

```python
sorted(list(enumerate(sim[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_idx=new_df[new_df['title']==movie].index[0]
    dis=sim[movie_idx]
    movie_list=sorted(list(enumerate(dis)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Iron Man')
```
